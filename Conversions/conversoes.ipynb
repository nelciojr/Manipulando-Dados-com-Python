```python
import ast
```

## Importando o arquivo ##

    Importando um arquivo dict em formato txt. Covertendo uma string contendo a lista em um objeto list.
  

```python
with open('candles.txt') as f:
    data = f.read()
    
current = ast.literal_eval(data)

#print(type(data))
#print(data)
print(type(current))
print(current)
```

```python
import pandas as pd
```

## Convertendo dict em dataframe ##


```python
useful_frame = pd.DataFrame.from_dict(current, orient='columns')
useful_frame = useful_frame.drop(columns=['id','volume'])
frames=[]
frames.append(useful_frame)
main = pd.concat(frames).drop_duplicates()
df = main.drop(columns=['at'])
df = df.loc[~df.index.duplicated()]
```

```python
df
```

```python
import numpy as np
from itertools import groupby
```

## Criando numpy array a partir de um dataframe ##
    
       Criando um nump array de um dataframe e convertendo os valores em simbolos de cores das velas.
    

```python
df.isnull().sum().sum()
df.fillna(method="ffill", inplace=True)
df = df.loc[~df.index.duplicated(keep = 'first')]
colorCandle = np.array([])
```

```python
for ind in df.index:
    if(df['open'][ind] > df['close'][ind]):
        colorCandle = np.append(colorCandle,'R')
    if(df['open'][ind] < df['close'][ind]):
        colorCandle = np.append(colorCandle,'G')
    if(df['open'][ind] == df['close'][ind]):
        colorCandle = np.append(colorCandle,'D')
```

```python
colorCandle
```

## Agrupando em lista de simbolos de cores ##
    
    Existem estratgias baseadas em cores de velas, pode-se us-las com a lista formada.
    

```python
listCandles = [list(groupCandles) for _, groupCandles in groupby(colorCandle)]
```
